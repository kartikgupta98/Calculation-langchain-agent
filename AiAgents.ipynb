```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model = 'llama-3.3-70b-versatile',
    temperature = 0,# controls the creativity of the model 0 to 1
    api_key = my_api_key
)
```

```python
response = llm.invoke('Write a 100 word essay on Data Analysis')
print(response.content)
```

```python
query = "What is 952124 * 123991? Give me numerical answer only."
response = llm.invoke(query)

print(response.content)
```

```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two integers together. Use this tool whenever you need to perform multiplication"""
    return a*b
```

```python
@tool
def divide(a:int, b:int) -> float:
    """Divides two integers. Use this tool whenever you need to perform division"""
    return a/b
```

```python
llm = llm.bind(parallel_tool_calls=False)
```

```python
from langgraph.prebuilt import create_react_agent

agent = create_react_agent(llm, tools=[multiply, divide])
```

```python
response = agent.invoke({"messages": [{"role":"user", "content": "What is 952124 * 123991. Now divide the result by 45 and give me final answer."}]})

print(response['messages'][-1].content)
```
